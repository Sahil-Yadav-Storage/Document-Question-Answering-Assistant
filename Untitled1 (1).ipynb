```python
import numpy as np
import torch
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
from transformers import pipeline

#Requirement 2 : Extract readable text from the PDF.
def pdf_text_extractor(pdf_path: str) -> str:
    """Extracting text from the inputted PDF file."""
    reader = PdfReader(pdf_path)
    extracted_text = [page.extract_text() for page in reader.pages if page.extract_text()]
    return "\n".join(extracted_text)

#Requirement 3 : Split the extracted text into chunks (recommended: 500 characters with 50-character overlap).
def chunk_text(text: str, chunk_size: int = 500, overlap: int = 50) -> list[str]:
    """Spliting text into chunks with chunk size as 500 characters and overlap of 50 characters."""
    chunks = []
    i = 0
    text_length = len(text)

    while i < text_length:
        end = min(i + chunk_size, text_length)
        chunk = text[i:end].strip()
        if chunk:
            chunks.append(chunk)
        i += (chunk_size - overlap)

    return chunks

class MemoryVectorStore:
    """We are using memory vector using cosine similarity with Numpy"""
    def __init__(self, embedding_model_name: str = "all-MiniLM-L6-v2"):
        self.model = SentenceTransformer(embedding_model_name)
        self.chunks = []
        self.embeddings = np.array([])

#Requirement 4 and 5 : Generate embeddings using any suitable embedding model and Store embeddings in memory.
    def embedding__indexing_chunks(self, chunks: list[str]):
        """Encoding chunks into embeddings and storing them in memory."""
        self.chunks = chunks

        temp_embeddings = self.model.encode(chunks, show_progress_bar=True)
        self.embeddings = np.array(temp_embeddings, dtype=np.float32)
        # Normalizing for faster searching
        normals = np.linalg.norm(self.embeddings, axis=1, keepdims=True)
        self.embeddings = self.embeddings / np.maximum(normals, 1e-8)
#Requirement 6 and 7 : Implement similarity search using cosine similarity with NumPy (no vector databases). Retrieve the top three relevant chunks for each user query.
    def similarity_search(self, query: str, top_k: int = 3) -> list[str]:
        """Retrieving top 3 chunks that match the query."""
        if len(self.chunks) == 0:
            return []

        query_embedding = self.model.encode([query])[0]
        query_norm = query_embedding / max(np.linalg.norm(query_embedding), 1e-8)

        similarities = np.dot(self.embeddings, query_norm)
        top_ans = np.argsort(similarities)[::-1][:top_k]

        return [self.chunks[idx] for idx in top_ans]


class DocumentQAAssistant:
    def __init__(self, vector_store: MemoryVectorStore, model_id: str = "Qwen/Qwen2.5-1.5B-Instruct", max_history_turns: int = 3):
        self.vector_store = vector_store
        self.max_history_turns = max_history_turns

        print(f"Loading local LLM ({model_id}) directly into Colab GPU. This may take a minute...")
        # Initialize Hugging Face pipeline for the open model
        self.pipe = pipeline(
            "text-generation",
            model=model_id,
            torch_dtype=torch.float16,
            device_map="auto"
        )

        # Keep system prompt separate to ensure it never gets deleted
        self.system_prompt = {
            "role": "system",
            "content": (
                "You are a helpful document assistant. Answer the user's questions "
                "strictly based on the retrieved context provided. If the answer cannot "
                "be inferred from the context, state that clearly."
            )
        }
        self.conversation_history = [self.system_prompt]

    def ask(self, query: str, top_k: int = 3) -> str:
        #Requirement 8 : Send only the retrieved chunks to the LLM and return the generated answer.
        retrieved_chunks = self.vector_store.similarity_search(query, top_k=top_k)
        context_str = "\n---\n".join(retrieved_chunks)

        # Injecting found context into the user's prompt
        prompt_with_context = f"Context Information:\n{context_str}\n\nQuestion: {query}"

        # Bonus requiremnet : Implement conversation history so the assistant can maintain context across multiple user queries.
        temp_history = self.conversation_history + [{"role": "user", "content": prompt_with_context}]

        # Generating responses
        outputs = self.pipe(
            temp_history,
            max_new_tokens=256,
            do_sample=False,
            return_full_text=False # Only gives assistant's answer
        )
        answer = outputs[0]["generated_text"].strip()

        # Updating memory
        self.conversation_history.append({"role": "user", "content": query})
        self.conversation_history.append({"role": "assistant", "content": answer})

        # Maintaining a conversation history of 3 prompt and answers as directed.
        max_allowed_messages = 1 + (self.max_history_turns * 2)

        while len(self.conversation_history) > max_allowed_messages:
            self.conversation_history.pop(1)
            self.conversation_history.pop(1)

        return answer

# Text to ensure you have reached the correct step, if not visible you have encountered an error
print("1. Extracting text from pdf ")
text = pdf_text_extractor(pdf_path)

print("2. Making chunks ")
chunks = chunk_text(text, chunk_size=500, overlap=50)

print("Making embeddings ")
# This is not an vector database but an inbuilt memory vector
vector_store = MemoryVectorStore()
vector_store.embedding__indexing_chunks(chunks)

# STARTING THE MODEL
assistant = DocumentQAAssistant(vector_store)
print("\n MODEL ONLINE. Please ask your questions")
```

```python
print("--- Hello! Please ask any question you have (Type 'exit' to quit the assistant) ---\n")

while True:
    question = input("You: ").strip()
    if not question:
        continue
    if question.lower() in ["exit"]:
        print("Ending assistant session.")
        break

    answer = assistant.ask(question, top_k=3)
    print(f"\nAssistant: {answer}\n" + "-" * 50)
```
